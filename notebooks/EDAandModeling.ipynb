```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to your data folder
sapce = pd.read_csv("C:\data-science-project\data\raw")
fmap = pd.read_csv("C:\data-science-project\data\raw")
cpi = pd.read_csv("C:\data-science-project\data\raw")

for name, df in [("SAPCE", sapce), ("FMAP", fmap), ("CPI", cpi)]:
    print(f"\n{name} dataset shape:", df.shape)
    display(df.head())
```

```python
for name, df in [("SAPCE", sapce), ("FMAP", fmap)]:
    print(f"\n{name} info:")
    df.info()
    print("\nSummary stats:")
    display(df.describe(include="all").T)
```

```python
for name, df in [("SAPCE", sapce), ("FMAP", fmap)]:
    print(f"\n{name} missing values (%):")
    print((df.isna().mean() * 100).round(2).sort_values(ascending=False))
```

```python
sapce_long = sapce.melt(
    id_vars=["GeoName", "Region", "Description", "Unit"],
    var_name="Year",
    value_name="Value"
)
sapce_long = sapce_long[sapce_long["Year"].str.isdigit()]
sapce_long["Year"] = sapce_long["Year"].astype(int)

# avg per year
sapce_long.groupby("Year")["Value"].mean().plot(marker="o")
plt.title("SAPCE: Average Expenditure per Year")
plt.xlabel("Year")
plt.ylabel("Mean Value(Millions USD)")
plt.show()
```

```python
fmap.groupby("Year")["Value"].mean().plot(kind="line", marker="o")
plt.title("FMAP: Average Value per Year")
plt.xlabel("Year")
plt.ylabel("Mean Value(Millions USD)")
plt.show()
```

```python
for name, df in [("SAPCE", sapce), ("FMAP", fmap)]:
    num_df = df.select_dtypes("number")
    corr = num_df.corr()
    print(f"\n{name} correlation matrix:")
    display(corr)
```
